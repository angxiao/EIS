```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
cells = ['25C01','25C02', '25C03','25C04', '25C05','25C06','25C07','25C08', '35C01', '35C02', '45C01', '45C02']

filenames = []

for cell in cells:
    file_path = f'EIS data/EIS_state_V_{cell}.txt'
    filenames.append(file_path)
```

```python
# Create a figure with 12 subplots (assuming a 3x4 grid)
fig, axs = plt.subplots(3, 4, figsize=(15, 12))

for i, filename in enumerate(filenames):
    df = pd.read_csv(filename, delimiter='\t')
    
    # Determine the subplot's position
    ax = axs[i//4, i%4]
    
    # Group by 'cycle number' and plot each group with a different color on the current subplot
    for name, group in df.groupby(df.columns[1]):
        ax.scatter(group.iloc[:, 2], group.iloc[:, -1], label=name)
    
    ax.set_xscale('log')
    ax.set_xlabel('freq/Hz')
    ax.set_ylabel('|Z|/Ohm')
    ax.set_title(f'Cell_{cells[i]}')
    ax.grid(True, which="both", ls="--", c='0.65')

# Adjust layout for better visualization
plt.tight_layout()
plt.show()
```
