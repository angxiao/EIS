```python
import numpy as np
import matplotlib.pyplot as plt
import GPy
import pandas as pd
from sklearn.metrics import r2_score
```

```python
TEST_CELLS = ['25C05', '25C06', '25C07', '25C08', '35C02', '45C02']
```

```python
filenames_common = ['EIS_data.txt', 'Capacity_data.txt']
data_common = {}
for fname in filenames_common:
    data_common[fname] = pd.read_csv(fname, header=None, delim_whitespace=True)

mean = data_common['EIS_data.txt'].mean(axis=0)
std = data_common['EIS_data.txt'].std(axis=0)
X_train = (data_common['EIS_data.txt'] - mean) / std
Y_train = data_common['Capacity_data.txt']

kernel = GPy.kern.RBF(input_dim=X_train.shape[1]) # Squared Exponential covariance function
model = GPy.models.GPRegression(X_train.values, Y_train.values, kernel)
model.optimize()
```

```python
# Set up a 2x3 grid of subplots
fig, axs = plt.subplots(2, 3, figsize=(10, 8))
axs = axs.ravel()

for idx, TEST_CELL in enumerate(TEST_CELLS):
    filenames = [f'EIS_data_{TEST_CELL}.txt', f'capacity{TEST_CELL}.txt']
    data = {}
    for fname in filenames:
        data[fname] = pd.read_csv(fname, header=None, delim_whitespace=True)

    X_test = (data[f'EIS_data_{TEST_CELL}.txt'] - mean) / std
    Y_test_cap, _ = model.predict(X_test.values)
    
# #     Compute the r2 score
#     actual_values = data[f'capacity{TEST_CELL}.txt'].values.squeeze()[:len(Y_test_cap)]
#     predicted_values = Y_test_cap.squeeze() / Y_test_cap[0]
#     r2 = r2_score(actual_values, predicted_values)

    axs[idx].plot(data[f'capacity{TEST_CELL}.txt'].values.squeeze() / data[f'capacity{TEST_CELL}.txt'].values[0], 'x',
                  color=[0/255, 130/255, 216/255], linewidth=3)
    axs[idx].plot(Y_test_cap.squeeze() / Y_test_cap[0], '+',
                  color=[205/255, 39/255, 70/255], linewidth=3)
    axs[idx].set_xlim([0, 300])
    axs[idx].set_ylim([0.3, 1.045])
    axs[idx].set_xlabel('Cycle Number', fontsize=15)
    axs[idx].set_ylabel('Capacity Retention', fontsize=15)
    axs[idx].set_title(f'{TEST_CELL}', fontsize=15)
    axs[idx].legend(['Measured', 'Estimated'], loc='upper right', fontsize=10, frameon=False)

#     # Display the r2 value on the subplot
#     axs[idx].text(10, 0.95, f'R2 = {r2:.2f}', fontsize=12, bbox=dict(facecolor='white', alpha=0.5))

plt.tight_layout()
plt.show()
```
