```python
import numpy as np
import matplotlib.pyplot as plt
import GPy
import pandas as pd
```

```python
TEST_CELL = '25C07'
```

```python
# Load relevant files
filenames = ['EIS_data.txt', 'Capacity_data.txt', f'EIS_data_{TEST_CELL}.txt', f'capacity{TEST_CELL}.txt']
data = {}
for fname in filenames:
    data[fname] = pd.read_csv(fname, header=None, delim_whitespace=True)
```

```python
# Training set for the GPR model
mean = data['EIS_data.txt'].mean(axis=0)
std = data['EIS_data.txt'].std(axis=0)
X_train = (data['EIS_data.txt'] - mean) / std
Y_train = data['Capacity_data.txt']
```

```python
# Gaussian Process Regression model
kernel = GPy.kern.RBF(input_dim=X_train.shape[1]) # Squared Exponential covariance function
model = GPy.models.GPRegression(X_train.values, Y_train.values, kernel)
model.optimize()
```

```python
# Testing set of the GPR model
X_test_35C02 = (data[f'EIS_data_{TEST_CELL}.txt'] - mean) / std

# Capacity estimation of the testing cell
Y_test_cap_35C02, Y_test_cap_35C02_var = model.predict(X_test_35C02.values)
```

```python
# Plotting the estimated capacity
plt.figure(figsize=(8, 6))
# f_upper = (Y_test_cap_35C02.squeeze() + np.sqrt(Y_test_cap_35C02_var.squeeze())) / Y_test_cap_35C02[0]
# f_lower = (Y_test_cap_35C02.squeeze() - np.sqrt(Y_test_cap_35C02_var.squeeze())) / Y_test_cap_35C02[0]
# plt.fill_between(f_lower, f_upper, color=[255/255, 191/255, 200/255])
plt.plot(data[f'capacity{TEST_CELL}.txt'].values.squeeze() / data[f'capacity{TEST_CELL}.txt'].values[0], 'x', color=[0/255, 130/255, 216/255], linewidth=3)
plt.plot(Y_test_cap_35C02.squeeze() / Y_test_cap_35C02[0], '+', color=[205/255, 39/255, 70/255], linewidth=3)
plt.xlim([0, 200])
plt.ylim([0.3, 1.045])
plt.xlabel('Cycle Number', fontsize=25)
plt.ylabel('Capacity Retention', fontsize=25)
plt.title(f'{TEST_CELL}', fontsize=25)
plt.legend(['Estimated', 'Measured'], loc='upper right', fontsize=25, frameon=False)
plt.show()
```
