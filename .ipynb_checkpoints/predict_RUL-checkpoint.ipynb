```python
import numpy as np
import matplotlib.pyplot as plt
import GPy
import pandas as pd
```

```python
TEST_CELL = '25C07'
```

```python
# Load relevant files
filenames = ['EIS_data.txt', 'RUL.txt', 'EIS_data_RUL.txt', 'EIS_data_35C02.txt', 'rul35C02.txt']
data = {}
for fname in filenames:
    data[fname] = np.loadtxt(fname)

# Training set of the GRP model
mean = np.mean(data['EIS_data.txt'], axis=0)
std = np.std(data['EIS_data.txt'], axis=0)
X_train_rul = (data['EIS_data_RUL.txt'] - mean) / std
Y_train_rul = data['RUL.txt'].reshape(-1, 1)  # reshaping to make it a column vector
```

```python
# Gaussian Process Regression model
kernel = GPy.kern.Linear(input_dim=X_train_rul.shape[1])
model = GPy.models.GPRegression(X_train_rul, Y_train_rul, kernel)
model.optimize()

# Testing set of the GPR model
X_test_rul_35C02 = (data['EIS_data_35C02.txt'] - mean) / std

# RUL prediction of the testing cell
Y_test_rul_35C02, Y_test_rul_35C02_var = model.predict(X_test_rul_35C02)
```

```python
# Plotting the predicted RUL vs. Actual RUL
plt.figure(figsize=(6, 4))
f_upper = Y_test_rul_35C02[:126] + np.sqrt(Y_test_rul_35C02_var[:126])
f_lower = Y_test_rul_35C02[:126] - np.sqrt(Y_test_rul_35C02_var[:126])
plt.fill_between(np.arange(0, 252, 2), f_lower[::-1].squeeze(), f_upper[::-1].squeeze(), color=[193/255, 221/255, 198/255])
plt.plot(data['rul35C02.txt'], Y_test_rul_35C02[:127], 'H', 
         color=[119/255, 172/255, 45/255], markersize=6, markerfacecolor=[119/255, 172/255, 45/255], linewidth=1)
plt.xlim([0, 252])
plt.ylim([0, 300])
plt.xlabel('Actual RUL', fontsize=25)
plt.ylabel('Predicted RUL', fontsize=25)
plt.title('Test cell', fontsize=25)
plt.grid(True)
plt.show()
```
