```python
import matplotlib.pyplot as plt
from capacity import CapacityData
import pandas as pd
```

```python
cells = ['25C01','25C02', '25C03','25C04', '25C05','25C06','25C07','25C08', '35C01', '35C02', '45C01', '45C02']

file_paths = []

for cell in cells:
    file_path = f'Capacity data/Data_Capacity_{cell}.txt'
    file_paths.append(file_path)
```

```python
new_df = pd.DataFrame()

# Create a 4x2 grid of subplots
fig, axes = plt.subplots(3, 4, figsize=(8, 6))

for i, (file_path, ax) in enumerate(zip(file_paths, axes.ravel())):
    capacity_data = CapacityData(file_path)
    cycle_numbers = capacity_data.df['cycle number'].values
    capacity_retention = capacity_data.df['capacity retention'].values

    # Plot data from CapacityData
    ax.plot(cycle_numbers, capacity_retention, 'x', color=[0/255, 130/255, 216/255], linewidth=3)
    
    # Plot the horizontal line at y=0.75
    ax.axhline(y=0.8, color='gray', linestyle='--')
    ax.set_xlim([0, 400])
    ax.set_ylim([0.2, 1.045])
    ax.set_xlabel('Cycle Number')
    ax.set_ylabel('Capacity Retention')
    ax.set_title(f'Cell_{cells[i]}')

plt.tight_layout()
plt.show()
```
